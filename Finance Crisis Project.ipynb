##### ** Creating a new empty DataFrame called returns, will contain the returns for each bank's stock. returns are typically defined by:**

```python
tickers=["BAC","C","GS","JPM","MS","WFC"]
tickers
```

```python
returns=pd.DataFrame()   #empty dataframe
```

```python
for x in tickers:
    returns[x+'Return'] = bank_stocks.xs('Close',axis=1,level=1)[x]
returns.head()
```

- using pandas pct_change() method on the Close column to create a column representing this return value.


```python
for x in tickers:
    returns[x+'Return'] = bank_stocks.xs('Close',axis=1,level=1).pct_change()
returns.head()
```

```python
for x in tickers:
    returns[x+'Return'] = bank_stocks.xs('Close',axis=1,level=1)[x].pct_change()
returns.head()
```

#### Moving Averages of Lows (all Banks)

moving averages for these stocks in the year 2008.


```python
bank_stocks.loc['2008-01-01':'2008-12-31'].xs('Low',axis=1,level=1)['BAC'].plot(figsize=(12,6),label='BofA Low')
rolling_avg = pd.DataFrame()
rolling_avg['30 Day Avg'] = bank_stocks.loc['2008-01-01':'2008-12-31'].xs('Low',axis=1,level=1)['BAC'].rolling(window=30).mean()
rolling_avg['30 Day Avg'].plot(figsize=(12,6),label='30 Day Avg')
plt.legend()
```
